```python
import matplotlib.pyplot as plt
import scipy.io as scio
import numpy as np
metric_values_iterations_dims = scio.loadmat(f'./results_save/metric_values_CalmRange5_CalmTrainNum10_1000.mat')#metric_values_pilot_calm_center.mat
metric_values = metric_values_iterations_dims['metric_values']
AUC = metric_values['AUC'][0,0]
all_colors = ['#2E94B9', "#fa625f", '#62C8A5', '#f29c2b']
plt.bar(1,AUC[500,0],width = 0.5, color = all_colors[0])
plt.bar(3,AUC[500,1],width = 0.5, color = all_colors[0])
plt.bar(5,AUC[500,2],width = 0.5, color = all_colors[0])
plt.bar(7,AUC[500,3],width = 0.5, color = all_colors[0], label = r'With linear layer')
metric_values_iterations_dims = scio.loadmat(f'./results_save/metric_values_without_linear_layer_1000.mat')#metric_values_pilot_calm_center.mat
metric_values = metric_values_iterations_dims['metric_values']
AUC = metric_values['AUC'][0,0]
all_colors = ['#2E94B9', "#fa625f", '#62C8A5', '#f29c2b']

plt.bar(1.5,AUC[500,0],width = 0.5, color = all_colors[1], label = r'Without linear layer')
plt.bar(3.5,AUC[500,1],width = 0.5, color = all_colors[1])
plt.bar(5.5,AUC[500,2],width = 0.5, color = all_colors[1])
plt.bar(7.5,AUC[500,3],width = 0.5, color = all_colors[1])
plt.legend(frameon=False)
plt.ylim((0.8,1))
plt.xticks((1.25,3.25,5.25,7.25),[r'$\beta=1$',r'$\beta=10$',r'$\beta=100$',r'$\beta=1000$'])
# plt.grid()
plt.xlabel('Iterations')
plt.ylabel('AUC value')
plt.savefig(f'../figures/auc_500_linear_layer.png',dpi=600)
plt.savefig(f'../figures/auc_500_linear_layer.pdf')
```
